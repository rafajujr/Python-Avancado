```python
# Um gerador que vai contar as palavra da linha e verificar se são iguas
def linhas_com_numeros_de_palavras(arquivo, numero):
    with open(arquivo, 'r', encoding='utf-8') as file:
        for linha in file:
            if len(linha.split()) == numero:
                yield linha
```

```python
arquivo = 'texto.txt'
numero = 5
for linha in linhas_com_numeros_de_palavras(arquivo, numero):
    print(linha)
```
