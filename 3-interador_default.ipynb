```python
import itertools
```

```python
for i in range(5):
    print(i)
```

```python
#Exemplo - Iterar uma lista 
frutas = ['maça', 'banana', 'laranja']
print(frutas)

for fruta in frutas:
    print(fruta)
```

```python
#Exemplo 3 - iterar um dicionário
telefones = {'João': '1234-5678', 'Maria': '9876-5432', 'Pedro': '5555-5555'}
for chave in telefones:
    print(f'{chave}: {telefones[chave]}')
```

```python
#Exemplo 4 - Enumerate()
for indice, fruta in enumerate(frutas):
    print(f'{indice}: {fruta}')
```

```python
#Exemplo 5 - Zip()
nomes = ['João', 'Maria', 'Pedro']
idades = [30,25,35]
for nome, idade in zip(nomes, idades):
    print(f'{nome} tem {idade} anos')
```

```python
#Exemplo 6 - Itertools
for i in itertools.count(start=10, step=2):
    if i > 20:
        break
    print(i)
```
