```python
class NomeComletra:
    def __init__(self, nomes, letra):
        self.nomes = nomes
        self.letra = letra
        self.indice = 0

    def __iter__(self):
        return self

    def __next__(self):
        while self.indice < len(self.nomes):
            nome = self.nomes[self.indice]
            self.indice += 1
            if nome.startswith(self.letra):
                return nome
            else:
                print(f"O nome {nome} não começa com a letra {self.letra}")
        raise StopIteration
```

```python
lista_nomes = ['Rafael', 'João', 'Maria', 'Pedro', 'Ana']
letra = 'A'
interador_nomes = NomeComletra(lista_nomes, letra)

for nome in interador_nomes:
    print(nome)
```
