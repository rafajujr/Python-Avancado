Metaclasse
- Cria classe de forma dinâmica.
- Pode ser útil na criação de frameworks onde as classes precisam ter um compotamento específico.

```python
class MeuMeta(type):
    def __new__(cls, nome, bases, dct):
        dct['novo_atributo'] = 'Valor adicionado pela metaclasse'
        return super().__new__(cls, nome, bases, dct)
```

```python
class MinhaClasse(metaclass=MeuMeta):
    pass
```

```python
obj = MinhaClasse()
obj.novo_atributo
```

```python
class ValidadorMeta(type):
    def __new__(cls, nome, bases, dct):
        Validacoes = dct.get('validacoes', {})

        for attr, tipo in Validacoes.items():
            if not callable(tipo):
                raise TypeError(f'O tipo de validação para o {attr} deve ser uma função.')
            
            # Add função de validação
            def valida_func(self, value, attr=attr, tipo=tipo):
                if not isinstance(value, tipo):
                    raise ValueError(f"{attr} deve ser do tipo {attr.__name___}")
                
            # Renomeia a função para evitar problemas de escopo
            valida_func.__name__ = f"set_{attr}"
            dct[f"set_{attr}"] = valida_func

        return super().__new__(cls, nome, bases, dct)
```

```python
class Usuario(metaclass=ValidadorMeta):
    validacoes = {
        'nome': str,
        'idade': int
    }

    def __init__(self, nome, idade):
        self.nome = nome
        self.idade = idade
```

```python
try:
    user = Usuario('Rafael', '20')
    print(f"O nome do usuario é: {user.nome} e a sua idade é: {user.idade}")

    user2 = Usuario("Juliana", user.idade)
    print(f"{user2.nome} tem a idade de {user2.idade}")
except TypeError as e:
    print(e)
except ValueError as e:
    print(e)
```
