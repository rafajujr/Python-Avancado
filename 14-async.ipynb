Async

- Permite que você escreva código que possa ser execultado de forma não bloqueante.
- Facilita a execoção de múltiplas operações simultâneas.

```python
import asyncio
```

```python
async def tarefa(nome, tempo):
    print(f"Tarefa {nome} iniciada")
    await asyncio.sleep(tempo)
    print(f"Tarefa {nome} concluida")
```

```python
async def main():
    await asyncio.gather(
        tarefa("Tarefa 1", 1),
        tarefa("Tarefa 2", 2),
        tarefa("Tarefa 3", 3)
    )
```

```python
try:
    asyncio.get_running_loop()
except RuntimeError:
    asyncio.run(main())

asyncio.ensure_future(main())
```
