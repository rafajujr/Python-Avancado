Type Hint

- É uma maneira de indicar o tipo de variáveis, parâmetros e valores de retorno de funções.

```python
def soma(a, b):
    return a + b
```

```python
def somaTypeHint(a: int, b: int) -> int:
    return a + b
```

```python
print(somaTypeHint(1, 2))
print(soma('1', '2'))
```

```python
from typing import List
```

```python
class Estudantes:
    def __init__(self, nome: str, idade: int, notas: List[float]) -> None:
        self.nome = nome
        self.idade = idade
        self.notas = notas

    def media(self) -> float:
        return sum(self.notas) / len(self.notas)
```

```python
def alunos_acima_da_media(alunos: list[Estudantes], media: float) -> list[str]:
    return [aluno.nome for aluno in alunos if aluno.media() > media]
```

```python
alunos = [
    Estudantes("Juliana", 20, [8, 9, 10]),
    Estudantes("Benjamim", 19, [8, 5, 10]),
    Estudantes("Rafael", 20, [8, 6, 8]),
    Estudantes("Ureia", 21, [8, 9, 2])
]
```

```python
resultados = alunos_acima_da_media(alunos, 7)
resultados
```

```python
media_alunos = [aluno.media() for aluno in alunos]
media_alunos
```
