Programação Funcional
- Trata a computação como a avaliação de funções matemáticas e evita estados e dados mutáveis

```python
# Função de alta ordem - é uma função que recebe outra fnção como argumento
def quadrado(x):
    return x * x

def aplicar_funcao(funcao, lista):
    return [funcao(x) for x in lista]
```

```python
numeros = [1,2,3,4,5]
resultado = aplicar_funcao(quadrado, numeros)
print(resultado)
```

```python
# Dados de exemplo
usuarios = [
    {"nome": "Juliana", "idade": 46},
    {"nome": "Benjamim", "idade": 12},
    {"nome": "Rafael", "idade": 39},
    {"nome": "Ureia", "idade": 30}
]
```

```python
def extrair_nome(usuarios):
    return list(map(lambda u: u['nome'], usuarios))

def filtrar_maiores_que(usuarios, idade):
    return list(filter(lambda u: u['idade'] > idade, usuarios))

def calcular_media_idade(usuarios):
    if not usuarios:
        return 0
    total_idades = sum(map(lambda u: u['idade'], usuarios))
    return total_idades / len(usuarios)
```

```python
lista_nomes = extrair_nome(usuarios)
maior18 = filtrar_maiores_que(usuarios, 18)
media = calcular_media_idade(maior18)

print(f"Os nomes dos ususraios da lista são: {lista_nomes}")
print(f"Na lista os que são de maior são: {usuarios}")
print(f"A média de idade da lista é de {media}")
```

```python
from functools import lru_cache
```

```python
@lru_cache(maxsize=None)
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)
```

```python
for i in range(10):
    print(f"Fibonacci({i}) =  {fibonacci(i)}")
```
