Zip
- Combinar mútlipos interáveis (lista ou tupla) em um único iterável

```python
nomes = ["Benjamim", "Juliana", "Rafael", "Ureia"]
idades = [12, 46, 39, 21]

for nome, idade in zip(nomes, idades):
    print(f"{nome} tem {idade} anos")
```

```python
pessoas = dict(zip(nomes, idades))
pessoas
```

```python
pessoas_mais_30 = {nome: idade for nome, idade in pessoas.items() if idade < 20}
pessoas_mais_30
```
